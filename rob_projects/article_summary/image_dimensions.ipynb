```python
from skimage.io import imread, imshow
img = imread('https://i.imgur.com/YobrZ8r.png')
dims = (1080, 1920)
ideal_width = dims[1]
ideal_height = dims[0]
ideal_aspect = ideal_width / ideal_height

size = img.shape

print('Original image: ', (0,0,size[0],size[1]))

width = size[0]
height = size[1]
aspect = width/height

if aspect > ideal_aspect:
    new_width = int(ideal_aspect * height)
    resize = (
        (0, 0, int(new_width), int(height)),
        (int(width-new_width), 0, int(width), int(height))
    )
else:
    new_height = int(width / ideal_aspect)
    resize = (
        (0, 0, int(width), int(new_height)), 
        (0, int(height-new_height), int(width), int(height))
    )

print('Red image: ', resize[0])

print('Grn image: ', resize[1])

# Red section
img[resize[0][0]:resize[0][2], resize[0][1]:resize[0][3], 0] = 180

# Green section
img[resize[1][0]:resize[1][2], resize[1][1]:resize[1][3], 1] = 180

imshow(img)
```
